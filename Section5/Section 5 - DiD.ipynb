```r
# Read in a .csv file containing data.
minwage_data <- read.csv('nj_min_wage.csv')
```

```r
# You can get some summary statistics for all the variables using the summary command
summary(minwage_data)
```

```r
# You can also do it for an individual variable. Here, we look at the starting wage, wage_st,
# and chain identifier, chain.
summary(minwage_data$wage_st)
summary(minwage_data$chain)
```

```r
# What if you want to look at summary statistics for a subset of observations?
# Let's compare summary statistics for Chain==1 and Chain ==2

summary(minwage_data[which(minwage_data$chain == 1),]$wage_st)
summary(minwage_data[which(minwage_data$chain == 2),]$wage_st)
summary(minwage_data[which(minwage_data$chain == 1 & minwage_data$shore == 0),]$wage_st)
```
